```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/NBA_All-Star_Game')[2] #getting the data from the website, third table only
```

```python
data
```

```python
#how to drop the missing data columns
data = data.dropna(how="any")
data
```

```python
#remove unnecessary columns
data = data.drop(['Host arena', 'Game MVP'], axis=1)
data
```

```python
#remove unnecessary information in the host city column
data['Host city'] = data['Host city'].str.rsplit(',').str[0] 
data
```

```python
#split the result into two new columns 
data[['East', 'West']] = data['Result'].str.split(',', 1, expand=True)
data
```

```python
#drop the result column
data = data.drop(['Result'], axis = 1)
```

```python
data
```

```python
data['East'] = data['East'].str.extract('(\d+)', expand=False) #keep only the numerical values
```

```python
data['West'] = data['West'].str.extract('(\d+)', expand=False)
```

```python
data = data.dropna(how="any") #drop missing values
data
```

```python
data['East'] = data['East'].astype(int) #convert results to integers
data['West'] = data['West'].astype(int)
data
```

```python
data['diff'] = data['East'] - data['West'] #create the difference in scores variable
```

```python
data
```

```python
grouped = data.groupby('diff') #group by score difference
grouped
```

```python
sized = grouped.size() #get the counts for each score difference
sized
```

```python
host = data.groupby('Host city').mean() #group by the host city
host
```

```python
len(host) #see how many cities there are
```

```python
host['counts'] = data['Host city'].value_counts() #create a column with city frequencies
host
```

```python
host = host.drop(['diff'], axis = 1) #drop the score difference
host
```

```python
host = host.sort_values(by='counts', ascending = False) #sort by the number of times each city hosted
host
```

```python
above2 = host['counts'] > 1 #keep only those with scores above 1
host[above2]
```
